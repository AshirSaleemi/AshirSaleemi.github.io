```python
# Importing libraries
from vega3 import VegaLite
import numpy as np
import pandas as pd
import geopandas as gpd
from sodapy import Socrata
import IPython
import altair as alt
from IPython.display import display
import seaborn as sns
df = pd.read_csv("CDPH_Environmental_Complaints.csv")
print(df.head())
print(df.info())
```

```python
# Cleaning the data

# Convert 'COMPLAINT DATE' to datetime data type
df['COMPLAINT DATE'] = pd.to_datetime(df['COMPLAINT DATE'], errors='coerce')

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values in 'STREET NUMBER TO' with 0
df['STREET NUMBER TO'].fillna(0, inplace=True)

# Drop rows with missing 'INSPECTOR' values
df.dropna(subset=['INSPECTOR'], inplace=True)

print(df.info())

df.to_csv('CDPH_Environmental_Complaints_Clean.csv', index=False)
```
